```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

## K-Means Clustering

```python
no_of_cluster = 4

blr_cluster = blr_venues.drop(['Categories', 'Neighborhood', 'Address', 'Code', 'Venue', 'Rating_Bin'], 1)
```

```python
kMeans = KMeans(n_clusters = no_of_cluster, random_state = 0).fit(blr_cluster)
```

```python
blr_venues_final = blr_venues
```

```python
blr_venues_final.insert(0, 'Cluster Label', kMeans.labels_)
blr_venues_final.head(5)
```

```python
import matplotlib.colors as colors
```

```python


# create map
map_clusters = folium.Map(location=[12.954294, 77.620335], zoom_start=13)

# set color scheme for the clusters
x = np.arange(no_of_cluster)
ys = [i + x + (i*x)**2 for i in range(no_of_cluster)]
colors_array = cm.brg(np.linspace(0, 1, len(ys)))
color_palette = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbor, venue, cluster in zip(blr_venues_final['Latitude'], blr_venues_final['Longitude'], blr_venues_final['Neighborhood'], blr_venues_final['Venue'], blr_venues_final['Cluster Label']):
    label = folium.Popup(str(venue) + str(neighbor) + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=color_palette[cluster-1],
        fill=True,
        fill_color=color_palette[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
